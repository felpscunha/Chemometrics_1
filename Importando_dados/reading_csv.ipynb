```python
import pandas as pd # biliboteca para manipulação de conjuntos de dados
import glob # biblioteca para ler os arquivos em um diretório
import matplotlib.pyplot as plt # visualização dos dados
```

```python
path = 'C:\\Users\\felip\\OneDrive\\Doutorado\\Dados_IV\\plan_1\\'
arquivos = [f for f in glob.glob(path + '*.csv', recursive=True)] # leitura de todos os arquivos .csv dentro do caminho especificado
```

```python
# organização dos dados // leitura de cada arquivo + concatenação em uma lista para depois transformar em um dataframe
array_df = []
for x in arquivos:
    temp_df = pd.read_csv(x, sep=';', decimal=',')
    temp_df = temp_df.drop('Criado como novo conjunto de dados', axis=1)
    temp_df = temp_df.T
    array_df.append(temp_df)
df = pd.concat(array_df, axis=0)
df = df.drop(columns=0)
columns = []
for x in range(3351): # corrigindo valores das colunas
    columns.append(4000 - x)
df.columns = columns
for col in df.columns:
    df[col] = df[col].str.replace(',', '.').astype(float) # corrigindo o formato dos arquivos para float
```

```python
fig, ax = plt.subplots(figsize=(20,10))
ax.plot(df.T)
plt.xlabel(r'Número de Onda ($\ cm^{-1}$ )', fontsize=15)
plt.ylabel('Absorbância', fontsize=15)
#plt.ylim([0.00, 0.06])
plt.xlim([4000, 650])
#plt.legend(fontsize=14)
plt.show()
```
