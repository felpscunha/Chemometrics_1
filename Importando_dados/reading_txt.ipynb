**Como importar dados em Python?**  

Esse é o primeiro passo para começar a trabalhar com os conjuntos de dados obtidos a partir de suas análises químicas. Quando analisamos alguma amostra em equipamentos como espectrofotômetros, cromatógrafos, etc, temos um grande número de dados. Isso se deve ao fato de que cada comprimento de onda ou tempo de retenção (em nossos exemplos) corresponde a uma variável de nosso futuro conjunto de dados.

Aqui vou dar um exemplo de como importar dados de uma amostra de infravermelho (FTIR - ATR) de um polímero de celulose. O arquivo está salvo em formato TXT com o seguinte formato.

arquivo txt.png

**Detalhes:**



1.   Temos algumas linhas de cabeçalho 
2.   Unidades de X e Y são número de onda (1/cm) e transmitância (%T) 
3.   1ª coluna: números de onda; 2ª coluna: transmitância em %.



```
# Finalmente vamos começar
# 1º Passo: Importar biblioteca(s) para fazer upload dos arquivos para o Colab
from google.colab import files

uploaded = files.upload()


```

```
# 2º Passo: Biblioteca Pandas para trabalhar com conjunto de dados (nesse caso nem tão necessário, pois só vamos usar uma amostra)
import pandas as pd 

amostra_1 = pd.read_csv('CBE1.txt', skiprows=4, sep='\t', index_col=0, names=['Número de onda (1/cm)', '%Transmitância']) # Criando uma variável que vai conter a amostra em formato de DataFrame do Pandas. 
# O comando read_csv leva o caminho para o arquivo (nesse caso já estamos com o arquivo CBE1.txt no próprio diretório do programa)
# skiprows=3 para pular as linhas de cabeçalho, sep= os dados estão separados por um espaço e não por vírgula, e eu disse que a coluna 0 é o índice dos meus dados
# names é para designar nomes melhores pras colunas
```

```
# Verificando se os dados foram importados corretamente 
amostra_1.head()
```

Dados importados! Agora vamos só dar uma olhada em como podemos representar esses dados espectrais em forma de gráfico

```
import matplotlib.pyplot as plt # Biblioteca padrão para plotar gráficos de forma simples

plt.plot(amostra_1) # Aqui apenas informei o conjunto de dados do pandas que eu havia criado e a figura foi criada
plt.xlabel('Número de onda (1/cm)') # Nome para o eixo X
plt.ylabel('Transmitância (%)') # Nome para o eixo Y
plt.show() # Plotar o gráfico efetivamente
```
