```python
import pandas as pd # bliblioteca para trabalhar com conjuntos de dados
import matplotlib.pyplot as plt # biblioteca para visualização de dados
```

```python
df = pd.read_excel('/content/GASOLINA2.xlsx', index_col=0) # leitura do conjunto de dados 
# necessário especificar o caminho pro seu arquivo
```

```python
df.head() # visualização das primeiras 5 linhas dos dados
```

```python
X = df.drop('% etanol', axis=1) # obtendo a matriz dos dados (exceto a coluna com valor da concentração de etanol)
X.head(1) 
```

```python
# plotando os dados espectrais 

plt.figure(figsize=(15,10)) # tamanho da figura

plt.plot(X.T) # matriz dos dados X transposta

plt.xlabel('Comprimento de onda (nm)') 
plt.ylabel('Absorbância')

plt.show() # mostrar a figura
```
